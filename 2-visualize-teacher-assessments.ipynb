```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
allAssessmentDF = pd.read_csv('../../astla-data/dart-preposttest/slate-data/1-teacherAssessments/all_teacherAssessments.csv', sep=',', encoding='utf-8') #.set_index('Unnamed: 0')

for var in allAssessmentDF.columns[1:]:
    allAssessmentDF[var] = allAssessmentDF[var].astype(int)
allAssessmentDF
```

```python
statsDF = pd.DataFrame()
statsDF['accuracy'] = allAssessmentDF.mean(axis=1)
statsDF['std'] = allAssessmentDF.std(axis=1)
statsDF['majority_vote'] = statsDF['accuracy'].apply(lambda x: x>0.5)
statsDF
```

```python
# Creating dataset
data = [list(statsDF['accuracy'])]

#Creating figure
fig = plt.figure()
 
# Creating plot
plt.boxplot(data)
plt.title("Accuracy scores per word")
plt.xticks([1],['Accuracy'])
plt.grid()
plt.ylabel("Accuracy")
 
# show plot
plt.show()
```
